```python
# Data fit transform in to x and y
X = df[["age", "sex", "bmi", "children", "smoker", "region"]]
y = df["charges"]
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# fit the scaler in the x_train and x_test
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
from sklearn.metrics import mean_squared_error, mean_absolute_error
def modelresult(prediction, y_test):
    print("mean_squared_error{}".format(mean_squared_error(y_test, prediction)))
    print("root_mean_squred_error on model is {}".format(np.sqrt(mean_squared_error(y_test, prediction))))
    print("mean_absolute_error on model is {}".format(mean_absolute_error(y_test, prediction)))

from sklearn.linear_model import LinearRegression
le = LinearRegression()
le.fit(scaler.transform(x_train), y_train)



```

```python
# predict the model
prediction = le.predict(scaler.transform(x_test))

modelresult(prediction, y_test)

```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
svrmodel =  SVR()
param_gridsvr = {"C": [0.001, 0.01, 0.1, 0.5, 1], "kernel": ['linear', 'poly', 'rbf'], "gamma": ["scale", "auto"], "degree": [2,3,4,5]}
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
gridsvr.fit(x_train, y_train)
print("Best Parameter for model is {}".format(gridsvr.best_params_))
```

```python
# Predict the model
prediction = gridsvr.predict(x_test)
modelresult(prediction, y_test)
```

```python
# Now we do this on Radom Forest
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
param_gridrf = {"bootstrap": [True], "max_depth": [5, 10, 15], "max_features": ["auto", "log2"], "n_estimators": [2,3,4,5,6,7,8,9,10]}
gridrf = GridSearchCV(model, param_gridrf)
gridrf.fit(x_train, y_train)

```

```python
prediction = gridrf.predict(x_test)

modelresult(prediction, y_test)
```
